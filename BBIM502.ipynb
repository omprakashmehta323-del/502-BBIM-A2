```python
# # File Name: langham_hotel_management
# Author: Gopal Jaishi
# Description: A hotel management system for LANGHAM Hotel 
# that allows staff to add, delete, allocate, 
# and manage room bookings without exception handling.
# Date: 02 01 2025
###############################################################

# Import os module for file operations like checking existence
try:
    import os  # for file operations
    from datetime import datetime  # for timestamping backup files
except ImportError as ie:
    print(f"Import_Error: Required module not found - {ie}")
    exit()  # Exit program if critical modules are missing

# Dictionary to store all room details globally
rooms = {}  

# -------------------------------------------------------------
# Function: add_room()
# Purpose: To add a new room to the hotel system.
# -------------------------------------------------------------
def add_room():
    """
    This function allows staff to add a new room to the hotel system.
    The user is prompted to enter the room number, type, and price.
    If the room already exists, the user is informed. The function
    handles invalid input, EOF errors, and unexpected exceptions.
    """
    try:
        # Prompt user to enter the room number
        room_number = input("Enter Room Number: ")
        # Prompt user to enter the room type
        room_type = input("Enter Room Type (Single/Double/Deluxe): ")
        # Prompt user to enter the room price, convert to float
        price = float(input("Enter Room Price per night: "))

        # Check if room already exists
        if room_number in rooms:
            print("Room already exists.")
        else:
            # Add the room details to the rooms dictionary
            rooms[room_number] = {
                "type": room_type,       # Store room type
                "price": price,          # Store price per night
                "allocated": False,      # Room initially not allocated
                "customer": ""           # No customer assigned yet
            }
            # Confirm successful addition
            print(f"Room {room_number} added successfully! Thanks")

    # Handle invalid numeric input for price
    except ValueError:
        print("Value_Error: Please enter a valid numeric value for the room price.")
    # Handle end-of-file error if input unexpectedly ends
    except EOFError:
        print("Error: No input detected. Please try again.")
    # Catch-all for any other unexpected errors
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# -------------------------------------------------------------
# Function: delete_room()
# Purpose: To delete an existing room from the system.
# -------------------------------------------------------------
def delete_room():
    """
    This function deletes an existing room from the hotel records.
    The user enters the room number to delete. If the room exists, it
    is removed; otherwise, the user is informed that the room is not found.
    Handles NameError and unexpected exceptions.
    """
    try:
        # Prompt user to enter room number to delete
        room_number = input("Enter Room Number to delete from list: ")
        # Check if the room exists
        if room_number in rooms:
            # Delete the room from dictionary
            del rooms[room_number]
            # Confirm deletion
            print(f"Room {room_number} deleted successfully!, Thanks")
        else:
            # Inform if room not found
            print("Room not found.")

    # Handle undefined variable 'rooms'
    except NameError:
        print("Name_Error: Variable 'rooms' is not defined.")
    # Catch-all for any other unexpected errors
    except Exception as e:
        print(f"Unexpected Error: {e}")

# -------------------------------------------------------------
# Function: display_rooms()
# Purpose: To display all available rooms with their details.
# -------------------------------------------------------------
def display_rooms():
    """
    This function displays all available rooms with details.
    For each room, it prints room number, type, price, and allocation status.
    Handles TypeError, IndexError, SyntaxError, and other exceptions.
    """
    try:
        # Check if the 'rooms' dictionary has any entries
        if rooms:
            # Print the table header for better readability
            print("Room No\tType\tPrice\tAllocated")
            # Loop through each room in the dictionary
            for rn, info in rooms.items():
                # Print the room number, type, price, and allocation status
                print(f"{rn}\t{info['type']}\t{info['price']}\t{info['allocated']}")
        else:
            # Inform the user if no rooms are available in the system
            print("Rooms are not available.")
    
    # Handle the case where the 'rooms' variable is of an invalid type
    except TypeError:
        print("Type_Error: Room data type is invalid.")
    
    # Handle the case where list or sequence indexing goes out of range
    except IndexError:
        print("Index_Error: Attempted to access invalid index in a list or sequence.")
    
    # Handle syntax errors if somehow introduced during execution
    except SyntaxError:
        print("Syntax_Error: There is a mistake in your code syntax.")
    
    # Catch-all for any other unexpected errors
    except Exception as e:
        print(f"Unexpected Error: {e}")

    

# -------------------------------------------------------------
# Function: allocate_room()
# Purpose: Allocate a room to a customer if available.
# -------------------------------------------------------------
def allocate_room():
    """
    This function allocates a room to a customer if available.
    The user enters a room number and customer name. If the room is not already allocated,
    the room is assigned to the customer and marked as allocated.
    Handles EOFError, NameError, and other unexpected exceptions.
    """
    try:
        # Prompt user to enter room number for allocation
        room_number = input("Enter Room Number to allocate: ")
        # Check if room exists
        if room_number in rooms:
            # Check if room is not already allocated
            if not rooms[room_number]["allocated"]:
                # Prompt user for customer name
                customer_name = input("Enter Customer Name: ")
                # Mark room as allocated
                rooms[room_number]["allocated"] = True
                # Store customer name
                rooms[room_number]["customer"] = customer_name
                # Confirm allocation
                print(f"Room {room_number} allocated to {customer_name}.")
            else:
                # Inform if room already allocated
                print("Room is already allocated.")
        else:
            # Inform if room does not exist
            print("Room not found.")
    except EOFError:
        print("EOF_Error: No input detected. Please try again.")
    except NameError:
        print("Name_Error: Variable 'rooms' is not defined.")
    except Exception as e:
        print(f"Unexpected Error: {e}")

# -------------------------------------------------------------
# Function: display_allocation()
# Purpose: Display all allocated rooms with customer details.
# -------------------------------------------------------------
def display_allocation():
    """
    This function displays all rooms that are currently allocated.
    It lists the room number, customer name, and room type for each allocated room.
    Handles TypeError and other unexpected exceptions.
    """
    try:
        # List of rooms that are currently allocated
        allocated_rooms = [rn for rn, info in rooms.items() if info["allocated"]]
        if allocated_rooms:
            # Print allocated rooms
            print("Allocated Rooms:")
            for rn in allocated_rooms:
                info = rooms[rn]
                print(f"{rn} -> {info['customer']} ({info['type']})")
        else:
            # Inform if no rooms are allocated
            print("No rooms are allocated currently.")
    except TypeError:
        print("Type_Error: Room data type is invalid.")
    except Exception as e:
        print(f"Unexpected Error: {e}")

# -------------------------------------------------------------
# Function: billing_deallocate()
# Purpose: Handle billing and free the room after checkout.
# -------------------------------------------------------------
def billing_deallocate():
    """
    This function handles billing and deallocates a room after checkout.
    The user enters the room number and number of nights stayed.
    Calculates the total bill and frees the room.
    Handles ValueError, ZeroDivisionError, OverflowError, TypeError, EOFError, and general exceptions.
    """
    try:
        # Prompt user for room number to checkout
        room_number = input("Enter Room Number to checkout: ")
        # Check if room exists and is allocated
        if room_number in rooms and rooms[room_number]["allocated"]:
            # Prompt user for number of nights stayed
            nights = int(input("Enter number of nights stayed: "))
            # Prevent division by zero for billing
            if nights == 0:
                raise ZeroDivisionError("Number of nights cannot be zero for billing.")
            # Calculate total bill
            bill = rooms[room_number]["price"] * nights
            # Check if bill exceeds float max limit
            if bill > 1e308:
                raise OverflowError("Bill calculation exceeds maximum float limit.")
            # Print bill amount
            print(f"Customer {rooms[room_number]['customer']} owes ${bill:.2f}")
            # Free the room
            rooms[room_number]["allocated"] = False
            rooms[room_number]["customer"] = ""
            print(f"Room {room_number} is now available for booking.")
        else:
            # Inform if room not allocated or does not exist
            print("Room is either not allocated or does not exist.")
    except ValueError:
        print("Value_Error: Please enter a valid number of nights.")
    except ZeroDivisionError as zde:
        print(f"ZeroDivision_Error: {zde}")
    except OverflowError as oe:
        print(f"Overflow_Error: {oe}")
    except TypeError:
        print("Type_Error: Invalid data type for price or nights.")
    except EOFError:
        print("EOF_Error: No input detected. Please try again.")
    except Exception as e:
        print(f"Unexpected Error: {e}")

# -------------------------------------------------------------
# Function: save_allocation_to_file()
# Purpose: Save all room data to a text file for record keeping.
# -------------------------------------------------------------
def save_allocation_to_file():
    """
    This function saves the current room allocation data to a text file.
    It writes details like Room Number, Type, Price, Allocation status, and Customer name.
    Handles IOErrors if the file cannot be written, and other unexpected exceptions.
    """
    try:
        filename = "LHMS_764706052.txt"
        # Open file in write mode
        with open(filename, "w") as file:
            if rooms:
                # Write room details to file
                for rn, info in rooms.items():
                    file.write(
                        f"Room: {rn}, Type: {info['type']}, Price: {info['price']}, "
                        f"Allocated: {info['allocated']}, Customer: {info['customer']}\n"
                    )
                print(f"Room data successfully saved to {filename}.")
            else:
                # Write message if no rooms exist
                file.write("No rooms available.\n")
                print("No data to save. Empty file created.")
    except IOError:
        print("IO_Error: Cannot write to file.")
    except Exception as e:
        print(f"Unexpected Error: {e}")


# -------------------------------------------------------------
# Function: display_allocation_from_file()
# Purpose: Read and display saved room allocation data from a file
# Handles: FileNotFoundError, IOError, Exception
# -------------------------------------------------------------
def display_allocation_from_file():
    """
    This function reads and displays saved room allocation data from a text file.
    Checks if the file exists and handles FileNotFoundError, IOError, and general exceptions.
    """
    try:
        filename = "LHMS_764706052.txt"  # Name of the file where room data is stored

        # Check if the file exists before attempting to read
        if os.path.exists(filename):
            # Open the file in read mode
            with open(filename, "r") as file:
                print("\n--- Room Allocation Records from File ---")  # Header for display
                content = file.read()  # Read the entire content of the file

                # If file is not empty, display content; otherwise, show message
                print(content if content else "File is empty.")
        else:
            # Inform user if the file does not exist
            print("File does not exist. Please save data first.")

    # Handle the case where file is missing (should be rare because we check os.path.exists)
    except FileNotFoundError as fnf:
        print(f"FileNotFound_Error: {fnf}")

    # Handle general Input/Output errors, e.g., permission issues
    except IOError:
        print("IO_Error: Cannot read file.")

    # Catch-all for any other unexpected exceptions
    except Exception as e:
        print(f"Unexpected Error: {e}")


# -------------------------------------------------------------
# Function: backup_and_clear_file()
# Purpose: Backup data to another file and clear original file.
# -------------------------------------------------------------
def backup_and_clear_file():
    """
    This function creates a backup of the main room allocation file
    and clears the original file. Backup file is timestamped.
    Handles IOError and other exceptions.
    """
    try:
        main_file = "LHMS_764706052.txt"
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_file = f"LHMS__764706052_Backup_{timestamp}.txt"
        # Check if main file exists
        if os.path.exists(main_file):
            # Read original file and append to backup
            with open(main_file, "r") as original, open(backup_file, "a") as backup:
                data = original.read()
                backup.write(data)
            # Clear the original file
            with open(main_file, "w") as f:
                f.write("")
            print(f"Data backed up to {backup_file} and main file cleared.")
        else:
            print("Main file not found. Nothing to back up.")
    except IOError:
        print("IO_Error: Cannot read/write file.")
    except Exception as e:
        print(f"Unexpected Error: {e}")

# -------------------------------------------------------------
# Function: main_menu()
# Purpose: Provides user-friendly menu-driven interface.
# -------------------------------------------------------------
def main_menu():
    while True:
        try:
            # Display main menu options
            print("\n--- Management System of LANGHAM Hotel ---")
            print("1. Add Room")
            print("2. Delete Room")
            print("3. Display Rooms Details")
            print("4. Allocate Room")
            print("5. Display Room Allocation Details")
            print("6. Billing & Deallocation")
            print("7. Save Room Allocation to File")
            print("8. Display Room Allocation from File")
            print("9. Backup and Clear Room Allocation File")
            print("0. Exit")

            # Prompt user for menu choice
            choice = input("Enter your choice: ")

            # Call corresponding function based on user choice
            if choice == "1":
                add_room()
            elif choice == "2":
                delete_room()
            elif choice == "3":
                display_rooms()
            elif choice == "4":
                allocate_room()
            elif choice == "5":
                display_allocation()
            elif choice == "6":
                billing_deallocate()
            elif choice == "7":
                save_allocation_to_file()
            elif choice == "8":
                display_allocation_from_file()
            elif choice == "9":
                backup_and_clear_file()
            elif choice == "0":
                # Exit the program
                print("Exiting: Thank You for Staying With Us!")
                break
            else:
                print("Invalid choice. Please enter a valid option.")
        except EOFError:
            print("EOF_Error: No input detected. Please try again.")
        except Exception as e:
            print(f"Unexpected Error: {e}")

# -------------------------------------------------------------
# Program Entry Point
# -------------------------------------------------------------
if __name__ == "__main__":
    try:
        main_menu()
    except ImportError:
        print("Import_Error: Required module not found.")
    except Exception as e:
        print(f"Unexpected Error: {e}")
```
